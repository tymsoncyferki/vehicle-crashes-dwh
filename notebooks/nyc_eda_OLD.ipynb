żeby nie zmieniać rozmiaru danych możemy zmergować tylko główny pojazd biorący udział w wypadku

```python
nyc_vehicles['VEHICLE_MAKE'].astype(str).value_counts()
```

```python
# weźmy jeden pojazd na jedną kolizję
nyc_vehicles['VEHICLE_MAKE'].fillna('', inplace=True)

def longest_name(series):
    longest = series.str.len().idxmax()
    return series.loc[longest]

result = nyc_vehicles.groupby('COLLISION_ID')['VEHICLE_MAKE'].agg(longest_name)
```

```python
result.reset_index()['VEHICLE_MAKE'].value_counts().head(50)
```

chujowy format, jeśli chcemy używać samochodów będzie potrzebne dużo ręcznej roboty (w dodatku większość to i tak nulle), plus prawie wszystkie modele to nulle :(
